```
import imgaug as ia
import imageio
import imgaug.augmenters as iaa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib
import matplotlib.pyplot as plt
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
from google.colab import files
```

```
img = files.upload()

for x in img.keys():
  image = x 
  print({image})
```

```
ac = imageio.imread("revolver.jpeg")
```

```
ia.imshow(ac)
```

```
bbs = BoundingBoxesOnImage([BoundingBox(x1=13, x2=268, y1=23, y2=151)], shape=ac.shape)
ia.imshow(bbs.draw_on_image(ac, size=5))
```

```

move=iaa.ElasticTransformation(alpha=50, sigma=5)
image_aug, bbs_aug = move(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move1=iaa.Flipud(0.5)
image_aug, bbs_aug = move1(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move2=iaa.GammaContrast((2.0, 2.0))
image_aug, bbs_aug = move2(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move3=iaa.EdgeDetect(alpha=(0, 0.7))
image_aug, bbs_aug = move3(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move4=iaa.Affine(shear=(-10, 10))
image_aug, bbs_aug = move4(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move5=iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)})
image_aug, bbs_aug = move5(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))

```

```
move6=iaa.Affine(rotate=(-40, 40))
image_aug, bbs_aug = move6(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```

```
move7=iaa.Affine(rotate=(-90, 90))
image_aug, bbs_aug = move7(image=ac, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(image_aug, size=5))
```
